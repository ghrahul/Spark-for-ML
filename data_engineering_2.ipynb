```python
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .appName("Python Spark data cleaning and engineering") \
    .config("spark.some.config.option", "some-value") \
    .getOrCreate()
```

```python
import pandas as pd
```

```python
listing = pd.read_csv("datasets/airbnb/listings.csv")
```

```python
listing
```

```python
listing.columns
```

```python
listing.to_parquet("datasets/airbnb/listing.parquet")
```

```python
file_location = "datasets/airbnb/listing.parquet"
file_type = "parquet"

# CSV options
infer_schema = "false"
first_row_is_header = "True"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df_listing = spark.read.format(file_type) \
  .option("multiline", "true") \
  .option("quote", "\"") \
  .option("escape", "\"") \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df_listing)
```

```python
df_listing.columns
```

```python
df_listing.select("price").distinct().show()
```

```python
from pyspark.sql.functions import udf, col
```

```python
from pyspark.sql.types import FloatType 

def trim_char(string):
    return string.strip("$")


spark.udf.register("trim_func", trim_char)
trim_func_udf = udf(trim_char)
```

```python
df_listing.select("property_type", "price", trim_func_udf("price").cast(FloatType()).alias("price_f")).groupBy("property_type").agg({"price_f":"average"}).show()
```

```python
import pyspark.sql.functions as f
from pyspark.sql.window import Window
df_neigh_count=df_listing.groupBy("neighbourhood").agg(({"neighbourhood":"count"})).withColumnRenamed("count(neighbourhood)", "count")
df_neigh_count.show()
```

```python
df_neigh_count = df_neigh_count.withColumn("percentage",f.col('count')/f.sum('count').over(Window.partitionBy()))
```

```python
df_neigh_count.select("neighbourhood").show(1000)
```

```python
file_location = "datasets/airbnb/neighbourhoods.csv"
file_type = "csv"

# CSV options
infer_schema = "True"
first_row_is_header = "True"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df_neighbourhoods = spark.read.format(file_type) \
  .option("multiline", "true") \
  .option("quote", "\"") \
  .option("escape", "\"") \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df_neighbourhoods)
```

```python
df_neighbourhoods.select("neighbourhood").show(1000)
```

```python
df_neighbourhoods.show()
```

```python
nei_group=df_neigh_count.alias("nc").join(df_neighbourhoods.alias("ne"),col("nc.neighbourhood") == col("ne.neighbourhood")).select("nc.*","ne.neighbourhood_group")

display(nei_group)
```

```python
from pyspark.sql.functions import isnan, when, count, col
df_listing.select([count(when(isnan(c) | col(c).isNull(), c)).alias("c") for c in df_listing.columns]).show()
```

```python
from pyspark.sql.types import DateType
from datetime import datetime

dfunc = udf(lambda x: datetime.strptime(x, "%Y-%m-%d"), DateType())
spark.udf.register("datec", dfunc)
```

```python
df_reviews.withColumn("datetime", dfunc(col("date"))).show()
```
